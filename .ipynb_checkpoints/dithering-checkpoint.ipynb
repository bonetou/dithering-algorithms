```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = Image.open('images/cat1.png').convert('L')
img2 = Image.open('images/cat2.png').convert('L')

#convertendo para 8bits
img1.convert('P').save('images/cat1_grayscale.png')
img2.convert('P').save('images/cat2_grayscale.png')

img1 = np.asarray(img1)/255.
img2 = np.asarray(img2)/255.
```

```python
def floyd_steinberg(img):
    limite = 0.25
    x, y = img.shape
    
    for i in range(x-1):
        for j in range(y-1):
            
            pixel = img[i][j]
            
            if(pixel < limite):
                erro = img[i][j] - 0
                img[i][j] = 0 # preto
                
            elif((pixel > limite) and (pixel <= limite + 0.25)):
                erro = img[i][j] - 0.25
                img[i][j] = 0.25
                
            elif((pixel > limite + 0.25) and (pixel <= limite + 0.5)):
                erro = img[i][j] - 0.5
                img[i][j] = 0.5
                
            else:
                erro = img[i][j] - 1
                img[i][j] = 1 # branco

            img[i+1][j] += (3/8) * erro
            img[i][j+1] += (3/8) * erro
            img[i+1][j+1] += erro / 4
            
    return img
```

```python
img1_floyd_steinberg = floyd_steinberg(img1)
img2_floyd_steinberg = floyd_steinberg(img2)
plt.imsave('images/floyd-steinberg/cat1_floyd_steinberg.png', img1_floyd_steinberg, cmap='gray')
plt.imsave('images/floyd-steinberg/cat12floyd_steinberg.png', img2_floyd_steinberg, cmap='gray')
```

```python
def modularizacao_aleatoria(img):
    
    x, y = img.shape
    limite = 0.25
    # varrendo sobre todos os pixels
    # esquerda para direita e de cima para baixo
    for i in range(x):
        for j in range(y):
            
            pixel = img[i][j] + np.random.uniform(0, 0.25)

            if(pixel < limite):
                img[i][j] = 0 # preto
                
            elif((pixel > limite) and (pixel <= limite + 0.25)):
                img[i][j] = 0.25
                
            elif((pixel > limite + 0.25) and (pixel <= limite + 0.5)):
                img[i][j] = 0.5
                
            else:
                img[i][j] = 1 # branco
    return img
```

```python
img1_noise = modularizacao_aleatoria(img1)
img2_noise = modularizacao_aleatoria(img2)

plt.imshow(img1_noise, 'gray')
plt.imsave('images/modularizacao/cat1_modularizacao_aleatoria.png', img1_noise, cmap='gray')

plt.imshow(img1_noise, 'gray')
plt.imsave('images/modularizacao/cat2_modularizacao_aleatoria.png', img2_noise, cmap='gray')
```
